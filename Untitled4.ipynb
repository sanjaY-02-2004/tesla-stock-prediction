```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Load the dataset
data = pd.read_csv('/content/TESLA.csv')
data = data[['Date', 'Close']]  # Using the 'Close' price for prediction

# Convert the Date column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
data.index = data['Date']

# Prepare data for LSTM model
# Normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1, 1))

# Create a data structure with 60 timesteps and 1 output
X_train = []
y_train = []
for i in range(60, len(scaled_data)):
    X_train.append(scaled_data[i-60:i, 0])
    y_train.append(scaled_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)

# Reshape data to be in 3D [samples, time steps, features] for LSTM input
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

# Build the LSTM model
model = Sequential()

model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=25))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=1, epochs=1)

# Predict future prices (next 1 year)
future_days = 365  # 1 year
last_60_days = scaled_data[-60:]
X_future = [last_60_days]

X_future = np.array(X_future)
X_future = np.reshape(X_future, (X_future.shape[0], X_future.shape[1], 1))

future_predictions = []

for _ in range(future_days):
    pred_price = model.predict(X_future)
    future_predictions.append(pred_price)

    # Reshape pred_price to match the required shape
    pred_price = np.reshape(pred_price, (1, 1, 1))

    # Update the input for the next prediction by removing the first timestep and adding the prediction
    X_future = np.append(X_future[:, 1:, :], pred_price, axis=1)

# Transform predictions back to original price scale
future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Validation step (correcting the error)
# Create validation data from the end of the dataset
val_data = scaled_data[-(60 + future_days):]  # Use a portion of the data for validation
X_val, y_val = [], []
for i in range(60, len(val_data)):
    X_val.append(val_data[i-60:i, 0])
    y_val.append(val_data[i, 0])

X_val, y_val = np.array(X_val), np.array(y_val)

# Check the shape of X_val to ensure it's not empty or incorrectly shaped
print(f'Shape of X_val: {X_val.shape}')
print(f'Shape of y_val: {y_val.shape}')

# If X_val is empty or has unexpected shape, print an error message
if X_val.size == 0 or X_val.ndim != 2:
    print("Error: X_val is either empty or has an incorrect shape.")
else:
    # Reshape validation data to 3D
    X_val = np.reshape(X_val, (X_val.shape[0], X_val.shape[1], 1))

    # Predict validation data
    val_predictions = model.predict(X_val)
    val_predictions = scaler.inverse_transform(val_predictions)

    # Plot the validation data
    train = data[:len(data) - len(val_predictions)]
    valid = data[len(data) - len(val_predictions):]
    valid['Predictions'] = val_predictions

    plt.figure(figsize=(16, 8))
    plt.title('Model')
    plt.xlabel('Date')
    plt.ylabel('Close Price USD ($)')
    plt.plot(train['Close'])
    plt.plot(valid[['Close', 'Predictions']])
    plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
    plt.show()

# Classify future predictions into price ranges
price_ranges = {'<500': 0, '500-1000': 0, '1000-1500': 0, '1500-2000': 0, '>2000': 0}

for price in future_predictions:
    if price < 500:
        price_ranges['<500'] += 1
    elif 500 <= price < 1000:
        price_ranges['500-1000'] += 1
    elif 1000 <= price < 1500:
        price_ranges['1000-1500'] += 1
    elif 1500 <= price < 2000:
        price_ranges['1500-2000'] += 1
    else:
        price_ranges['>2000'] += 1

# Plot pie chart of price ranges
plt.figure(figsize=(8, 8))
plt.pie(price_ranges.values(), labels=price_ranges.keys(), autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'orange', 'red', 'purple'])
plt.title('Predicted Stock Price Ranges for the Next 1 Year')
plt.show()
```
